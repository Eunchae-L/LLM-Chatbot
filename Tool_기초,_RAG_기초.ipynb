````python
# app.py ì—ì„œ ì§ì ‘ ìˆ˜ì •í•˜ê² ìŠµë‹ˆë‹¤.
import streamlit as st
from openai import OpenAI
import requests
import json

API_KEY = ""
weather_api_key= ""

st.title("ë‚ ì”¨ë´‡ğŸŒ¤ï¸")
st.write("Hello â˜ï¸ğŸ¯KAIROS. ê¶ê¸ˆí•œ ë‚ ì”¨ ì •ë³´ë¥¼ ë¬¼ì–´ë³´ì„¸ìš”")

client = OpenAI(api_key=API_KEY)

if "openai_model" not in st.session_state:
  st.session_state.openai_model = "gpt-3.5-turbo"

# Initialize chat history
if "messages" not in st.session_state:
  st.session_state.messages = [{
    "role": "system",
    "content": """
    ë„ˆëŠ” ìœ ì €ì˜ íŠ¹ì • ì§€ì—­ì˜ ë‚ ì”¨ ì •ë³´ë¥¼ ë¬»ëŠ” ì§ˆë¬¸ì— ë‹µë³€í•˜ëŠ” ì±—ë´‡ì´ì•¼.
    í•œêµ­ì–´ë¡œ ì…ë ¥ë°›ì€ ë„ì‹œì´ë¦„ì„ ì˜ì–´ë¡œ ë°”ê¿”ì„œ ì²˜ë¦¬í•´ì¤˜.
    ë„ˆì—ê²ŒëŠ” ë‹¤ìŒê³¼ ê°™ì€ Toolì´ ìˆì–´.
    -OpenWeatherMap APIë¥¼ í™œìš©í•˜ì—¬ íŠ¹ì • ì§€ì—­ì˜ ë‚ ì”¨ ì •ë³´ë¥¼ ê°€ì ¸ì˜¤ëŠ” íˆ´
    í•¨ìˆ˜ì— ì—°ê²°í•  ê°’ì„ ê°€ì •í•˜ì§€ ë§ˆì„¸ìš”.
    ì‚¬ìš©ìì˜ ìš”ì²­ì´ ëª¨í˜¸í•œ ê²½ìš° ì„¤ëª…ì„ ìš”ì²­í•˜ì„¸ìš”
    í•œêµ­ì–´ë¡œ ì¹œì ˆí•˜ê²Œ ì´ëª¨ì§€ë¥¼ ì„ì–´ì„œ ë‹µë³€í•´ì¤˜
    """
  }]

def get_current_weather(location, format):
  # OpenWeatherMap API í‚¤ë¥¼ 'your_api_key_here' ìë¦¬ì— ì‹¤ì œ í‚¤ë¡œ êµì²´í•˜ì„¸ìš”
  api_key = weather_api_key
  # í˜„ì¬ ë‚ ì”¨ ì •ë³´ë¥¼ ìœ„í•œ OpenWeatherMap API ì—”ë“œí¬ì¸íŠ¸
  base_url = "http://api.openweathermap.org/data/2.5/weather?"

  # API ìš”ì²­ì„ ìœ„í•œ ì˜¨ë„ ë‹¨ìœ„ ì‹œìŠ¤í…œì„ í¬ë§·ì— ë”°ë¼ ë³€í™˜
  units = 'metric' if format == 'celsius' else 'imperial'

  # íŒŒë¼ë¯¸í„°ë¥¼ í¬í•¨í•˜ì—¬ ì™„ì„±ëœ API URL êµ¬ì„±
  complete_url = f"{base_url}appid={api_key}&q={location}&units={units}"

  # API ìš”ì²­ ìˆ˜í–‰
  response = requests.get(complete_url)
  print(response)
  # ìš”ì²­ì´ ì„±ê³µì ì´ì—ˆëŠ”ì§€ í™•ì¸
  if response.status_code == 200:
    # ì‘ë‹µ JSON íŒŒì‹±
    data = response.json()
    result = json.dumps(data)

    return result
  else:
    # ê°€ëŠ¥í•œ ì˜¤ë¥˜ ì²˜ë¦¬(ì˜ˆ: ìœ„ì¹˜ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŒ)
    return "ë‚ ì”¨ ì •ë³´ë¥¼ ê°€ì ¸ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ìœ„ì¹˜ë¥¼ í™•ì¸í•˜ê³  ë‹¤ì‹œ ì‹œë„í•˜ì„¸ìš”."

tools = [
  {
    "type": "function",
    "function": {
      "name": "get_current_weather",
      "description": "ë‚ ì”¨ ì •ë³´(e.g. ì˜¨ë„(temperature), ìŠµë„(humidity)ë¥¼ ê°€ì ¸ì˜¤ëŠ” í•¨ìˆ˜",
      "parameters": {
        "type": "object",
        "properties": {
          "location": {
            "type": "string",
            "description": "ë„ì‹œ ì´ë¦„ì€ ë¬´ì¡°ê±´ ì˜ì–´ê³ , ëŒ€ë¬¸ìë¡œ ì‹œì‘í•´ì•¼ í•œë‹¤.. e.g. ì„œìš¸ -> Seoul, ë¶€ì‚° -> Busan.",
          },
          "format": {
            "type": "string",
            "enum": ["celsius", "fahrenheit"],
            "description": "The temperature unit to use. Infer this from the users location.",
          },
        },
        "required": ["location", "format"],
      },
    }
  }]

# Display chat messages from history on app rerun
for message in st.session_state.messages:
  if message["role"] != "system":
    with st.chat_message(message["role"]):
      st.markdown(message["content"])

# Accept user input for chat
if prompt := st.chat_input("ë‚ ì”¨ì— ëŒ€í•´ ë¬¼ì–´ë³´ì„¸ìš”"):
  # Add user message to chat history
  st.session_state.messages.append({"role": "user", "content": prompt})

  # Display user message in chat message container
  with st.chat_message("user"):
    st.markdown(prompt)

  # Display assistant response in chat message container
  with st.chat_message("assistant"):
    message_placeholder = st.empty()
    full_response = ""
    response = client.chat.completions.create(
        model=st.session_state.openai_model,
        messages=st.session_state.messages,
        tools=tools,
    )
    finish_reason = response.choices[0].finish_reason
    if finish_reason == "tool_calls":
      # 6. Function call ì´ë¼ë©´ í•¨ìˆ˜ë¥¼ ì‹¤í–‰í•˜ê¸°
      response_message = response.choices[0].message
      arguments_string = response_message.tool_calls[0].function.arguments
      arguments = json.loads(arguments_string)
      function_name = response_message.tool_calls[0].function.name

      function_response = eval(f"{function_name}(**arguments)")

      # 7. ì‹¤í–‰ëœ ê²°ê³¼ë¥¼ LLMì—ê²Œ ì•Œë ¤ì£¼ê³  ë‹µë³€ì„ ìƒì„±í•˜ê¸°
      st.session_state.messages.append({
          "role": "function",
          "name": "get_current_weather",
          "content": function_response
      })

      second_response = client.chat.completions.create(
          model=st.session_state.openai_model,
          messages=st.session_state.messages,
          tools=tools,
      )

      final_message = second_response.choices[0].message.content

    elif finish_reason == "stop":
      final_message = response.choices[0].message.content

    message_placeholder.markdown(final_message)
  st.session_state.messages.append({"role": "assistant", "content": final_message})
````

````python
# LLMì˜ ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ì— ì œê³µí•  ë°ì´í„°ë² ì´ìŠ¤ ì •ë³´ë¥¼ ì¶”ì¶œí•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ê² ìŠµë‹ˆë‹¤. (skipí• ê¹Œ?)
def get_table_names(conn):
    """Return a list of table names."""
    table_names = []
    tables = conn.execute("SELECT name FROM sqlite_master WHERE type='table';")
    for table in tables.fetchall():
        table_names.append(table[0])
    # del table_names[0]
    return table_names


def get_column_names(conn, table_name):
    """Return a list of column names."""
    column_names = []
    columns = conn.execute(f"PRAGMA table_info('{table_name}');").fetchall()
    for col in columns:
        column_names.append(col[1])
    return column_names


def get_database_info(conn):
    """Return a list of dicts containing the table name and columns for each table in the database."""
    table_dicts = []
    for table_name in get_table_names(conn):
        columns_names = get_column_names(conn, table_name)
        table_dicts.append({"table_name": table_name, "column_names": columns_names})
    return table_dicts
````

````python
# LLMì—ê²Œ ì œê³µí•  tool í•¨ìˆ˜ë¥¼ ì •ì˜í•´ë³´ê² ìŠµë‹ˆë‹¤.

import json

def ask_database(query):
    """Function to query SQLite database with a provided SQL query."""
    try:
        results = str(conn.execute(query).fetchall())
    except Exception as e:
        results = f"query failed with error: {e}"
    return results

def execute_function_call(message):
    if message.tool_calls[0].function.name == "ask_database":
        query = json.loads(message.tool_calls[0].function.arguments)["query"]
        results = ask_database(conn, query)
    else:
        results = f"Error: function {message.tool_calls[0].function.name} does not exist"
    return results
````
